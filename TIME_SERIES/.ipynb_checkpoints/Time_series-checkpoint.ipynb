# Task 1. Arima approach

```python
tmp = df[['Close']]
```

```python
#!pip install statsmodels
```

```python
import statsmodels.api as sm 
import matplotlib.pyplot as plt           # for plots and graphics
import statsmodels.tsa.api as smt
%matplotlib inline
def tsplot(y, lags=None, figsize=(12, 7), style='bmh'):
    """
        Plot time series, its ACF and PACF, calculate Dickey–Fuller test
        
        y - timeseries
        lags - how many lags to include in ACF, PACF calculation
    """
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
        
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        
        y.plot(ax=ts_ax)
        p_value = sm.tsa.stattools.adfuller(y)[1]
        ts_ax.set_title('Time Series Analysis Plots\n Dickey-Fuller: p={0:.5f}'.format(p_value))
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax)
        plt.tight_layout()
```

```python
tsplot(tmp['Close'])
tmp.keys()
```

```python
tsplot(tmp.Close.diff(7)[7:])
```

```python
from statsmodels.tsa.stattools import acf

# Create Training and Test
train = tmp.Close[:1200]
test = tmp.Close[1200:]
```

```python
# Build Model
import statsmodels.api as sm 
model = sm.tsa.statespace.SARIMAX(train, order=(4, 1, 5), 
                                  seasonal_order=(1, 1, 0, 20))  
fitted = model.fit(disp=False)  
# fc, se, conf = fitted.forecast(len(test), alpha=0.05)  # 95% conf
res = model.fit(disp=False)
print(res.summary())
```

```python
# Forecast
fc = fitted.forecast(len(test), alpha=0.05)  # 95% conf
```

```python
# Make as pandas series
fc_series = pd.Series(fc.values, index=test.index)
```

```python
# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
# Plot residual errors
fitted.plot_diagnostics(figsize=(15,8))
plt.show()
```

```python
# Accuracy metrics
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    mae = np.mean(np.abs(forecast - actual))    # MAE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    return({'mape':mape, 'mae': mae, 
            'rmse':rmse})

forecast_accuracy(fc, test.values)
```

```python
import tensorflow as tf
```

```python
df = yf.download('AMD', start='2017-01-01', end='2022-03-10', progress=False,) # завантажуємо дані для AMD
df.tail()
```

```python
df.shape
```

```python
temp = df['Close'] # вибираємо величину, яку будемо передбачати
temp.plot()
```

```python
def df_to_X_y(df, window_size=5): # функція, що перетворює датасет у тривимірний
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
window = 15 # вікно необхідне в числових рядах для врахування попередніх значень (аналогія з контекстом для тексту)
X_w, y_w = df_to_X_y(temp, window)
X_w.shape, y_w.shape
```

```python
X_train, y_train = X_w[:1200], y_w[:1200] # розділяємо датасет на тренувальний, тестувальний та валідаційний
X_val, y_val = X_w[1200:1220], y_w[1200:1220]
X_test, y_test = X_w[1220:], y_w[1220:]
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

# будуємо модель з використанням бібліотеки tensorflow
model = Sequential()
model.add(InputLayer((15, 1)))
model.add(LSTM(32))
model.add(Dense(10, 'relu')) 
model.add(Dense(1, 'linear'))

model.summary()
```

```python
cp = ModelCheckpoint('model/', save_best_only=True)
model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.01), metrics=[RootMeanSquaredError()])
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=25, callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model = load_model('model/') 
```

```python
# Метрики для перевірки якості отриманої моделі
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # Mean absolute percentage error
    mae = np.mean(np.abs(forecast - actual))    # Mean absolute error
    rmse = np.mean((forecast - actual)**2)**.5  # Root Mean Square Error
    return({'mape':mape, 'mae': mae, 
            'rmse':rmse})
```

```python
train_predictions = model.predict(X_train).flatten()
train_results = pd.DataFrame(data={'Train predictions':train_predictions, 'Train actuals':y_train})
train_results
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train predictions'])
plt.plot(train_results['Train actuals'])
```

```python
forecast_accuracy(train_results['Train predictions'], train_results['Train actuals'])
```

```python
val_predictions = model.predict(X_val).flatten()
val_results = pd.DataFrame(data={'Validation predictions':val_predictions, 'Validation actuals':y_val})
val_results
```

```python
plt.plot(val_results['Validation predictions'])
plt.plot(val_results['Validation actuals'])
```

```python
forecast_accuracy(val_results['Validation predictions'], val_results['Validation actuals'])
```

```python
test_predictions = model.predict(X_test).flatten()
test_results = pd.DataFrame(data={'Test predictions':test_predictions, 'Test actuals':y_test})
test_results
```

```python
plt.plot(test_results['Test predictions'])
plt.plot(test_results['Test actuals'])
```

```python
forecast_accuracy(test_results['Test Predictions'], val_results['Actuals'])
```
